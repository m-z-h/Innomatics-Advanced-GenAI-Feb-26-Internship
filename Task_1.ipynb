```python
# 1. User Login Check
# Real-World Application: Authentication systems

def user_login_check(input_username, input_password):
    username = "admin"
    password = "1234"

    if input_username == username and input_password == password:
        return "Login Successful"
    else:
        return "Invalid Credentials"

# Test Case 1
result = user_login_check("admin", "1234")
print(f"Result: {result}")

# Test Case 2
result = user_login_check("MZH", "1235")
print(f"Result: {result}")
```

```python
# 2. Pass / Fail Analyzer
# Real-World Application: Academic evaluation systems

def pass_fail_analyzer(marks):
    pass_count = 0
    fail_count = 0

    for mark in marks:
        if mark >= 35:
            pass_count += 1
        else:
            fail_count += 1

    return pass_count, fail_count

marks = [45, 78, 90, 33, 60, 85, 25,75, 44, 24, 19]
pass_students, fail_students = pass_fail_analyzer(marks)

print(f"Marks analyzed: {marks}")
print(f"Total Students: {len(marks)}")
print(f"Passed Students: {pass_students}")
print(f"Failed Students: {fail_students}")
```

```python
# 3. Simple Data Cleaner
# Real-World Application: Data preprocessing before analysis

def simple_data_cleaner(names):
    cleaned_names = [name.strip().lower() for name in names]
    return cleaned_names

names = [" Alice ", "bob", " CHARLIE " , " MZH Plays"]
cleaned = simple_data_cleaner(names)

print(f"Original Names: {names}")
print(f"Cleaned Names: {cleaned}")
```

```python
# 4. Message Length Analyzer
# Real-World Application: Text filtering and validation systems

def message_length_analyzer(messages):
    for message in messages:
        length = len(message)
        status = "⚠️ Message longer than 10 characters" if length > 10 else "OK"
        print(f"Message: '{message}' | Length: {length} | Status: {status}")

messages = ["Hi", "Welcome to the MZH Plays", "OK"]
message_length_analyzer(messages)
```

```python
# 5. Error Message Detector
# Real-World Application: Monitoring and log analysis systems

def error_message_detector(logs):
    return logs.count("ERROR")

logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error_count = error_message_detector(logs)

print(f"Logs analyzed: {logs}")
print(f"Total ERROR messages found: {error_count}")
```

```python
# Problem 2: Search Query Keyword Analysis
# Submitted by: MD ZAHID HUSSAIN

import string

# Real Indian search query example
search_query = "Buy mobile phone online buy smartphone India best mobile buy phone India smartphone"

# Convert to lowercase
search_query = search_query.lower()

# Remove punctuation
search_query = search_query.translate(str.maketrans('', '', string.punctuation))

# Split words
words = search_query.split()

# Count frequency
word_count = {}

for word in words:
    word_count[word] = word_count.get(word, 0) + 1

# Filter words appearing more than once
result = {word: count for word, count in word_count.items() if count > 1}

print("Keywords searched more than once:")
print(result)
```

```python
# Problem 3: Sensor Data Validation
# Submitted by: MD ZAHID HUSSAIN

# Sensor readings (10 real readings)
sensor_readings = [12, 15, 18, 21, 24, 30, 33, 42, 50, 55]

valid_readings = []

# Check even readings
for index, value in enumerate(sensor_readings):
    if value % 2 == 0:
        valid_readings.append((index, value))

# Display result
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
# Problem 4: Email Domain Usage Analysis
# Submitted by: MD ZAHID HUSSAIN

emails = [
    "amit.sharma@gmail.com",
    "priya.singh@yahoo.com",
    "rahul.verma@gmail.com",
    "neha.gupta@outlook.com",
    "arjun.patel@gmail.com",
    "sneha.reddy@yahoo.com",
    "vikram.malhotra@gmail.com",
    "pooja.mehta@outlook.com",
    "rohit.kumar@gmail.com",
    "mdzahidhussain@gmail.com"
]

domain_count = {}

# Count domains
for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

# Calculate percentage
total = len(emails)

print("Email Domain Usage Percentage:")
for domain, count in domain_count.items():
    percentage = (count / total) * 100
    print(f"{domain}: {percentage:.0f}%")
```

```python
# Problem 5: Sales Spike Detection
# Submitted by: MD ZAHID HUSSAIN

# Daily sales (10 days, Indian retail example in ₹)
sales = [12000, 15000, 18000, 25000, 14000, 30000, 17000, 35000, 16000, 40000]

# Calculate average
average = sum(sales) / len(sales)

# Threshold = 30% above average
threshold = average * 1.30

print(f"Average Sales: {average:.2f}")
print("Sales Spike Days:")

# Detect spikes
for day, value in enumerate(sales, start=1):
    if value > threshold:
        print(f"Day {day}: ₹{value}")
```

```python
# Problem 6: Duplicate User ID Detection
# Submitted by: MD ZAHID HUSSAIN

user_ids = [
    "amit123",
    "priya456",
    "rahul789",
    "amit123",
    "neha321",
    "rahul789",
    "rohit654",
    "mdzahid786",
    "priya456",
    "amit123"
]

user_count = {}

# Count occurrences
for user in user_ids:
    user_count[user] = user_count.get(user, 0) + 1

# Display duplicates
print("Duplicate User IDs:")
for user, count in user_count.items():
    if count > 1:
        print(f"{user} → {count} times")
```

## Task 1: Conceptual Understanding

### Q1. What is the difference between "Love" and "love" in NLP?
**Answer:**  
In NLP, "Love" and "love" are treated as different tokens because of case sensitivity.  
This can increase vocabulary size unnecessarily.  
To avoid this, text is usually converted to lowercase.

---

### Q2. What happens if stopwords are not removed?
**Answer:**  
If stopwords are not removed, the dataset contains many common words like "is", "the", "am".  
This increases noise and may reduce model performance and efficiency.

---

### Q3. Provide two real-world scenarios where removing stopwords can be harmful.
**Answer:**  
1. In sentiment analysis, removing words like "not" can completely change meaning.  
   Example: "not good" → "good" (wrong meaning)

2. In chatbots or question answering, removing words like "what", "how" can break the sentence structure.

---

### Q4. Explain the difference between stemming and lemmatization.
**Answer:**  
Stemming cuts words to their base form using simple rules, which may not always be meaningful.  
Example: "running" → "run"

Lemmatization converts words into their correct base form using context and vocabulary.  
Example: "better" → "good"  
It is more accurate than stemming.

#Task 2: Preprocessing Function

```python
import re

def preprocess_text(text):
    # Handle empty or invalid input
    if not isinstance(text, str) or text.strip() == "":
        return [], ""

    # Convert to lowercase
    text = text.lower()

    # Remove URLs and emails
    text = re.sub(r'https?://\S+|www\.\S+|\S+@\S+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Reduce repeated characters (soooo → soo)
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)

    # Remove emojis / special characters
    text = re.sub(r'[^a-z\s]', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Tokenization
    tokens = text.split()

    # Remove small words except important ones
    tokens = [word for word in tokens if len(word) > 2 or word in ['no', 'not']]

    cleaned_sentence = " ".join(tokens)

    return tokens, cleaned_sentence
```

#Task 3: Stress Testing

```python
test_sentences = [
    "Bro I got this for just 999!!! 😭🔥",
    "Yaar this movie was sooo boringgg...",
    "Check this out: https://amazon.in deal is crazy!",
    "Call me maybe at 9876543210 later",
    "THIS COURSE IS ACTUALLY GOOD!!!",
    "Nooooo I don't like this at all",
    "Ok ok I understand now",
    "Win money $$$ fast now!!!",
    "I am not happy with the service",
    "Delivery was late but product is decent 🙂"
]

for s in test_sentences:
    tokens, clean = preprocess_text(s)
    print("Original:", s)
    print("Tokens:", tokens)
    print("Cleaned:", clean)
    print("-" * 50)
```

#Task 4: Token Analytics

```python
def analyze(tokens):
    if len(tokens) == 0:
        return 0, 0, 0

    total = len(tokens)
    unique = len(set(tokens))
    avg_len = sum(len(w) for w in tokens) / total

    return total, unique, avg_len

for s in test_sentences:
    tokens, _ = preprocess_text(s)
    t, u, a = analyze(tokens)

    print(s)
    print(f"Total Tokens: {t}, Unique Tokens: {u}, Avg Length: {a:.2f}")
    print("-" * 50)
```

### Analysis

**1. Which sentence had the most noise?**  
"Win money $$$ fast now!!!"  
It contains symbols, repeated punctuation, and promotional noise.

**2. Which sentence retained the most meaningful tokens?**  
"I am not happy with the service"  
Important sentiment words like "not" and "happy" are preserved.

#Task 5: Frequency Analysis

```python
from collections import Counter

all_tokens = []

for s in test_sentences:
    tokens, _ = preprocess_text(s)
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 Frequent Words:", counter.most_common(10))
print("Least Frequent Words:", counter.most_common()[-5:])
```

#Task 6: Full Pipeline

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

result = full_pipeline(test_sentences)
print(result)
```

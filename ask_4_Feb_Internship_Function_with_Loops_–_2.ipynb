```python
# Problem 1: Mobile Recharge Validation System

def validate_number(number):
    if not number.isdigit():
        print("Mobile number must contain digits only.")
        return False
    if len(number) != 10:
        print("Mobile number must be exactly 10 digits.")
        return False
    return True


providers = {
    1: ("Jio", [199, 299, 399, 599]),
    2: ("Airtel", [199, 299, 399, 599]),
    3: ("Vi", [199, 299, 399, 599]),
    4: ("BSNL", [147, 199, 397, 599])
}


# Step 1: Enter mobile number
while True:
    number = input("Enter mobile number: ").strip()
    if validate_number(number):
        break


# Step 2: Select provider
while True:
    print("\nSelect Provider:")
    for key, value in providers.items():
        print(f"{key}. {value[0]}")

    try:
        choice = int(input("Enter choice: "))
        provider_name, plan_list = providers[choice]
        break
    except (ValueError, KeyError):
        print("Invalid provider selection.")


# Step 3: Select plan
while True:
    print(f"\nAvailable plans for {provider_name}:")
    for plan in plan_list:
        print(f"₹{plan}")

    try:
        selected_plan = int(input("Select plan: ₹"))
        if selected_plan in plan_list:
            print(f"\nRecharge successful for {number} ({provider_name}) with ₹{selected_plan}")
            break
        else:
            print("Invalid plan. Select from the list.")

    except ValueError:
        print("Enter numbers only.")
```

```python
# Problem 2: Inventory Reorder Alert System

def check_inventory(inventory):
    for product, stock in inventory.items():
        if stock < 15:
            print(f"{product}: Reorder Alert (Stock = {stock})")
        else:
            print(f"{product}: Stock OK (Stock = {stock})")

inventory = {}

# Handle number of products input
while True:
    try:
        n = int(input("Enter number of products: "))
        break
    except ValueError:
        print("Please enter a valid number for the count.")

for i in range(n):
    product = input(f"\nEnter product {i+1} name: ")

    # Handle stock input
    while True:
        try:
            stock = int(input(f"Enter stock for {product}: "))
            inventory[product] = stock
            break
        except ValueError:
            print("Please enter a valid number for the stock level.")

print("\nInventory Status:")
check_inventory(inventory)
```

```python
# Problem 3: Student Result Processing System

def process_result(marks):
    total = 0

    for mark in marks:
        total += mark

    average = total / len(marks)

    if average >= 50:
        return f"Pass (Average = {average:.2f})"
    else:
        return f"Fail (Average = {average:.2f})"


marks = []

n = int(input("Enter number of subjects: "))

for i in range(n):
    mark = float(input(f"Enter marks for subject {i+1}: "))
    marks.append(mark)

result = process_result(marks)

print("\nResult:", result)
```

```python
# Problem 4: Cab Fare Estimator with Retry Option

def calculate_fare(distance, peak):
    fare = 50 + (12 * distance)

    if peak in ["yes", "y"]:
        fare += fare * 0.25

    return fare


while True:
    try:
        distance = float(input("Enter distance in km: "))

        # Validate peak input
        while True:
            peak_input = input("Is it peak hour? (yes/no): ").strip().lower()
            if peak_input in ["yes", "y", "no", "n"]:
                break
            print("Invalid input. Enter yes/y or no/n only.")

        total_fare = calculate_fare(distance, peak_input)
        print(f"Total Cab Fare: ₹{total_fare:.2f}")

        # Retry logic
        while True:
            retry = input("Calculate again? (yes/no): ").strip().lower()

            if retry in ["yes", "y"]:
                break   # restart main loop

            elif retry in ["no", "n"]:
                print("Fare calculation ended.")
                break   # break retry loop

            else:
                print("Invalid input. Enter yes/y or no/n only.")

        if retry in ["no", "n"]:
            break   # end program properly

    except ValueError:
        print("Invalid input. Enter numeric distance only.")
```

```python
# Problem 5: Employee Attendance Eligibility Checker

def check_eligibility(attendance_list):
    present_count = 0

    for day in attendance_list:
        if day == "P":
            present_count += 1

    percentage = (present_count / len(attendance_list)) * 100

    if percentage >= 75:
        return f"Eligible ({percentage:.2f}%)"
    else:
        return f"Not Eligible ({percentage:.2f}%)"


attendance = []

total_days = int(input("Enter total number of working days: "))

for i in range(total_days):
    while True:
        status = input(f"Enter attendance for day {i+1} (P/A): ").strip().upper()

        if status in ["P", "A"]:
            attendance.append(status)
            break
        else:
            print("Invalid input. Enter only P or A.")

result = check_eligibility(attendance)

print("Attendance Status:", result)
```

```python
# Problem 6: Password Strength Checker

def check_password(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"

    if len(password) < 8:
        return "Weak Password: Minimum length must be 8 characters"

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True

    if not has_digit:
        return "Weak Password: Must contain at least one digit"

    if not has_special:
        return "Weak Password: Must contain at least one special character (@#$)"

    return "Strong Password"


user_password = input("Enter password: ")

result = check_password(user_password)

print(result)
```

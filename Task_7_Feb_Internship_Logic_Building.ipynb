```python
# Problem 1: Smart Light Controller

class SmartLight:

    def __init__(self, name):
        self.name = name
        self.status = "OFF"

    def turn_on(self):
        self.status = "ON"

    def turn_off(self):
        self.status = "OFF"

    def display_status(self):
        print(f"{self.name} is {self.status}")


# User Input
light_name = input("Light Name: ")
action = input("Action (ON/OFF): ").strip().upper()

light = SmartLight(light_name)

if action == "ON":
    light.turn_on()
elif action == "OFF":
    light.turn_off()

light.display_status()
```

```python
# Problem 2: Employee ID Card System

class Employee:

    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_id_card(self):
        print("\nEmployee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")


# User Input
name = input("Employee Name: ")
emp_id = input("Employee ID: ")
department = input("Department: ")

employee = Employee(name, emp_id, department)
employee.display_id_card()
```

```python
# Problem 3: Mobile Contact Record

class Contact:

    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display_contact(self):
        print("\nContact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")


# User Input
name = input("Contact Name: ")
phone = input("Phone Number: ")

contact = Contact(name, phone)
contact.display_contact()
```

```python
# Problem 4: Product Price Tag Generator

class Product:

    def __init__(self, name, price):
        self.name = name
        self.price = price

    def display_price_tag(self):
        print(f"\nProduct: {self.name}")
        print(f"Price: ₹{self.price}")


# User Input
product_name = input("Product Name: ")
price = float(input("Price: "))

product = Product(product_name, price)
product.display_price_tag()
```

```python
# Problem 5: Movie Rating Display System

class Movie:

    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display_movie(self):
        print(f"\nMovie: {self.name}")
        print(f"Rating: {self.rating} / 5")


# User Input
movie_name = input("Movie Name: ")
rating = float(input("Rating: "))

movie = Movie(movie_name, rating)
movie.display_movie()
```

```python
# Problem 6: Delivery Address Manager

class Delivery:

    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    def display_delivery(self):
        print("\nDelivery Details")
        print(f"Customer: {self.customer_name}")
        print(f"Address: {self.address}")


# User Input
customer_name = input("Customer Name: ")
address = input("Address: ")

delivery = Delivery(customer_name, address)
delivery.display_delivery()
```
